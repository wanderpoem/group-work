```python
import pandas as pd
```

```python
gdp = pd.read_excel('sum.xlsx', sheet_name='Sheet1')
pop = pd.read_excel('sum.xlsx', sheet_name='Sheet2')
birth = pd.read_excel('sum.xlsx', sheet_name='Sheet3')
```

```python
gdp
```

```python
years = [i for i in range(2005, 2024)] 
data = []

for year in years:
    temp_df = pd.DataFrame({
        '地区': gdp['地区'],
        'GDP': gdp[year],  
        '人口': pop[year],
        '出生率': birth[year],
        '年份': year
    })
    data.append(temp_df)

df = pd.concat(data, ignore_index=True)
df
```

```python
province = df['地区'].unique()
tem = {}
for i in province:
    tem[i] = float(df[df['地区'] == i]['GDP'].values[0])
tem
```

```python
import json


with open('gdp_data_2005.json', 'w', encoding='utf-8') as f:
    json.dump(tem, f, ensure_ascii=False, indent=4)
```

```python
for j in range(2006,2025):
    tem = {}
    for i in province:
        tem[i] = float(df[df['地区'] == i]['GDP'].values[j-2005])
    with open(f'gdp_data_{j}.json', 'w', encoding='utf-8') as f:
        json.dump(tem, f, ensure_ascii=False, indent=4)
```

```python
import json

with open('angle2.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

for province, angles in data.items():
    for key in ['x轴', 'y轴', 'z轴', 'distance']:
        if key in angles:
            if key == 'distance':
                angles[key] = round(float(angles[key]), 2)
            else:
                angles[key] = round(-180 + float(angles[key]), 2)

with open('angle.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print("已生成 angle.json")
```

```python
years = [i for i in range(2005, 2024)]
provinces = gdp['地区'].unique()

for province in provinces:
    result = []
    for year in years:
        entry = {
            "年份": year,
            "GDP": float(gdp[gdp['地区'] == province][year].values[0]),
            "人口": float(pop[pop['地区'] == province][year].values[0]),
            "出生率": float(birth[birth['地区'] == province][year].values[0])
        }
        result.append(entry)
    filename = f"{province}_data.json"
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(result, f, ensure_ascii=False, indent=4)
    print(f"已生成 {filename}")
```
